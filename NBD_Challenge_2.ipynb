```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
alpha_list = [2.01, 2.05, 2.25]
b_list = list((np.array(alpha_list)-1)/np.array(alpha_list))
rho_list = np.linspace(0.5, 0.9, num=100)
lambda_list = list(np.array(rho_list*11))
mu_2 = 10
mu_1 = 1

list_opt_th_list = []
for a, b in zip(alpha_list, b_list):
    opt_th_list = []
    opt_E_D_list = []

    for lamb in lambda_list:

        theta_lower = ((mu_2*(a-1))/((b**a)*a*lamb))**(1/(1-a))
        th_low = (max(b,theta_lower))

        th_upp = ((1/(b**(a-1)))-((mu_1*(a-1))/((b**a)*a*lamb)))**(1/(1-a))

        th_list = np.linspace(th_low, th_upp, num=100)[1:-1]

        E_D_list = []
        for th in th_list:

            P_S2 = (b/th)**a
            P_S1 = 1-P_S2

            lamb_1 = P_S1*lamb
            lamb_2 = P_S2*lamb

            E_L1 = ((a*(b**a))*(((th**(1-a))-(b**(1-a)))))/(P_S1*(1-a))
            E_L1sq = ((a*(b**a))*((th**(2-a))-(b**(2-a))))/(P_S1*(2-a))

            E_L2 = ((a*(b**a))*((th**(1-a))))/(P_S2*(a-1))
            E_L2sq = ((a*(b**a))*(th**(2-a)))/(P_S2*(a-2))

            E_S1 = ((E_L1)/mu_1)+((lamb_1*(E_L1sq))/((2*mu_1)*(mu_1-((lamb_1*E_L1)))))
            E_S2 = ((E_L2)/mu_2)+((lamb_2*(E_L2sq))/((2*mu_2)*(mu_2-((lamb_2*E_L2)))))


            E_D = E_S1 + E_S2
            E_D_list.append(E_D)


        min_E_D = min(E_D_list)
        ind = E_D_list.index(min(E_D_list))
        opt_th = th_list[ind]
        opt_th_list.append(opt_th)
        opt_E_D_list.append(min_E_D)
    plt.plot(rho_list, opt_th_list, label = "a = " + str(a))

plt.xlabel('ρ', fontsize = 16)
# Set the y axis label of the current axis.
plt.ylabel('θ*', fontsize = 16)
# Set a title of the current axes.
plt.title('θ* as function of "ρ" for different alpha ', fontsize = 16)
plt.legend(loc = "upper left") 
plt.xticks()
plt.show
```
